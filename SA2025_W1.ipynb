### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here


median_mpg = Cars['mpg'].median()

above_median_mpg = Cars[Cars['mpg'] > median_mpg]
above_median_mpg['hp_per_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Step 4: Find the max ratio
max_ratio = above_median_mpg['hp_per_weight'].max()


best_cars = above_median_mpg[above_median_mpg['hp_per_weight'] == max_ratio]

print(best_cars)


```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here

import seaborn as sns
import matplotlib.pyplot as plt


mpg_trend = Cars.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()


plt.figure(figsize=(10, 6))
sns.lineplot(data=mpg_trend, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()


plt.show()

```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here

import seaborn as sns
import matplotlib.pyplot as plt

plot_data = Cars.dropna(subset=['horsepower', 'weight', 'mpg', 'origin'])


plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=plot_data,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    palette='Set1',
    hue_order=['japan', 'europe', 'usa'],
    sizes=(40, 200),
    alpha=0.7
)


plt.title('Horsepower vs. Weight by Origin (Size = MPG)', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()


plt.show()

```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here


origin_years = Cars.groupby('origin')['model_year'].nunique()
multi_year_origins = origin_years[origin_years > 1].index

filtered_cars = Cars[Cars['origin'].isin(multi_year_origins)].copy()
origin_stats = filtered_cars.groupby('origin').agg(
    mpg_std=('mpg', 'std'),
    appearances=('mpg', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

consistent_origins = origin_stats[
    (origin_stats['mpg_std'] < 1.0) & (origin_stats['appearances'] >= 2)
]

sorted_consistent_origins = consistent_origins.sort_values(
    by=['appearances', 'avg_mpg'], ascending=[False, False]
)

print(sorted_consistent_origins[['origin', 'appearances', 'avg_mpg']])

```
